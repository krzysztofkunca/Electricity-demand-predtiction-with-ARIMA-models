## Extracting

```python
#weather_path = '../Additional Data/Weather/'

from zipfile import ZipFile 

# specifying the zip file name 
#file_name = "2010_100_s.zip"
#file = weather_path + file_name

def extractFile(file_name):
        # opening the zip file in READ mode 
    with ZipFile(file_name, 'r') as zip: 
        # printing all the contents of the zip file 
#         zip.printdir() 

        # extracting all the files 
        log = ('Extracting all the files from: ' +file_name) 
        zip.extractall() 
        log = log + (' ... Done!') 
        
        print(log)
        return None
# test:
# extractFile("2010_100_s.zip")
```

```python
# Main loop

start_year = 2010
end_year   = 2020

for year in range(start_year, end_year+1):
    if(year != 2021):
        for ending in endings:
    #         print(str(year)+'_'+str(ending)+'_s.zip')
            file_name = str(year)+'_'+str(ending)+'_s.zip'
            try:
                extractFile(file_name)
            except:
                print('No such file ' + file_name)
    else:
        for i  in range(1,12+1):
            if(i >=10):
                file_name = str(year) +'_' +str(i) +'_s.zip'
            else: 
                file_name = str(year) +'_0' +str(i) +'_s.zip'
            try:
                extractFile(file_name)
            except:
                print('No such file: '+file_name)
```
